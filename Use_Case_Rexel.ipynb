# Pre-rocessing with pandas

```python
X_train = pd.get_dummies(data)
X_test = pd.get_dummies(test_data)
```

```python
X_train
```

```python
X_train.columns.equals(X_test.columns)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.columns.difference(X_test.columns)
```

```python
X_test[X_train.columns.difference(X_test.columns)[0]] = 0
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_test = X_test[X_train.columns]
```

#  Classification Trees (CART)

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=2)
```

```python
decision_tree.fit(X_train, y)
```

```python
export_graphviz(decision_tree, feature_names=X_train.columns)
```

```python
from IPython.display import Image
Image("tree.png", unconfined=True)
```

```python
decision_tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
decision_tree.fit(X_train, y)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, decision_tree.predict(X_test))
```

To avoid overfitting: max_depth=10

```python
decision_tree = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=10)
decision_tree.fit(X_train, y)
metrics.mean_absolute_error(y_test, decision_tree.predict(X_test))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=1000, criterion='entropy', n_jobs=4, max_depth=10)
random_forest.fit(X_train, y)
```

```python
metrics.mean_absolute_error(y_test, random_forest.predict(X_test))
```

###  Cross Validation

```python
from sklearn.cross_validation import StratifiedKFold
from sklearn.grid_search import GridSearchCV

cv = StratifiedKFold(y, n_folds=4)

params = {"max_depth": [3, 5, 7]}
rf = RandomForestClassifier(n_estimators=500)
estimator = GridSearchCV(rf, param_grid=params, verbose=2)
estimator.fit(X_train, y)
```
